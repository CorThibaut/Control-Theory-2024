```python
import numpy as np

import matplotlib.pyplot as plt
from IPython.display import display, clear_output
```

# Generation of data

```python
tMax = 10
N = 1000
f0 = 1
alpha = 0.1

t = np.linspace(0, tMax, N)
y1 = np.sin(2*np.pi*f0*t)
y2 = np.cos(2*np.pi*f0*t)
y3 = np.exp(alpha*t)
y4 = np.sqrt(t)
```

# Plot data

#### Example 1: not recommended

```python
plt.plot(t, y1, label = 'Sine')
plt.plot(t, y2, label = 'Cosine')
plt.xlabel('Time [s]')
plt.ylabel('Value of sine')
plt.title('Sine')
plt.xlim([0, tMax])
plt.legend()
plt.grid()
plt.show()
```

#### Example 2: not recommended

```python
plt.subplot(211)
plt.plot(t, y1, label = 'Sine')
plt.plot(t, y2, label = 'Cosine')
plt.ylabel('Value of sine')
plt.title('Sine')
plt.xlim([0, tMax])
plt.legend()
plt.grid()


plt.subplot(212)
plt.plot(t, y3, label = 'Exponential')
plt.xlabel('Time [s]')
plt.ylabel('Value of exponential')
plt.title('Exponential')
plt.xlim([0, tMax])
plt.legend()
plt.grid()

plt.subplots_adjust(hspace=0.4)

plt.show()
```

#### Example 3: recommended

```python
fig, ax = plt.subplots()

ax.plot(t,y1,'b-', linewidth=2, label='Sine')
ax.plot(t,y2,'g-', linewidth=2, label='Cosine')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Value of sine and cosine')
ax.set_title('Sine and cosine')
ax.set_xlim([0, tMax])
ax.grid()
ax.legend(loc='best');
```

#### Example 4: recommended

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(12)
fig.set_figwidth(12)

fig.delaxes(ax4)

ax1.plot(t,y1,'b-', linewidth=2, label='Sine')
ax1.plot(t,y2,'g-', linewidth=2, label='Cosine')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Value of sine and cosine')
ax1.set_title('Sine and cosine')
ax1.set_xlim([0, tMax])
ax1.grid()
ax1.legend(loc='best')

ax2.plot(t,y3,'m-', linewidth=2, label='Exponential')
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Value of exponential')
ax2.set_title('Exponential')
ax2.text(6.1, 1.1, 'alpha =  {}'.format(alpha), color="magenta", fontsize=10)
ax2.set_xlim([0, tMax])
ax2.grid()
ax2.legend(loc='best')

ax3.plot(t,y4,'c-', linewidth=2, label='Square root')
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Value of square root')
ax3.set_title('Square root')
ax3.set_xlim([0, tMax])
ax3.grid()
ax3.legend(loc='best')

fig.tight_layout(pad = 2.0)
```

#### Example 5: create plots upfront and set data later

```python
# Create figure upfront

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(12)
fig.set_figwidth(12)

fig.delaxes(ax4)

l1, = ax1.plot([],[],'b-', linewidth=2, label='Sine')
l2, = ax1.plot([],[],'g-', linewidth=2, label='Cosine')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Value of sine and cosine')
ax1.set_title('Sine and cosine')
ax1.grid()
ax1.legend(loc='best')

l3, = ax2.plot([],[],'m-', linewidth=2, label='Exponential')
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Value of exponential')
ax2.set_title('Exponential')
ax2.set_xlim([0, tMax])
ax2.grid()
ax2.legend(loc='best')

l4, = ax3.plot([],[],'c-', linewidth=2, label='Square root')
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Value of square root')
ax3.set_title('Square root')
ax3.set_xlim([0, tMax])
ax3.grid()
ax3.legend(loc='best')

# Set data

l1.set_data(t,y1)
l2.set_data(t,y2)
l3.set_data(t,y3)
l4.set_data(t,y4)

ax1.set_xlim([0, tMax])
ax1.set_ylim([-1.2, 1.2])

ax2.set_xlim([0, tMax])
ax2.set_ylim(0, np.max(y3))

ax3.set_xlim([0, tMax])
ax3.set_ylim(0, np.max(y4))

fig.tight_layout(pad = 2.0)
```

#### Example 6: set data as they become available; useful for real-time plots

```python
# Create figure upfront

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(12)
fig.set_figwidth(12)

fig.delaxes(ax4)

l1, = ax1.plot([],[],'b-', linewidth=2, label='Sine')
l2, = ax1.plot([],[],'g-', linewidth=2, label='Cosine')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Value of sine and cosine')
ax1.set_title('Sine and cosine')
ax1.grid()
ax1.legend(loc='best')

l3, = ax2.plot([],[],'m-', linewidth=2, label='Exponential')
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Value of exponential')
ax2.set_title('Exponential')
ax2.set_xlim([0, tMax])
ax2.grid()
ax2.legend(loc='best')

l4, = ax3.plot([],[],'c-', linewidth=2, label='Square root')
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Value of square root')
ax3.set_title('Square root')
ax3.set_xlim([0, tMax])
ax3.grid()
ax3.legend(loc='best')

# Continuous update of the data

t = []
y1 = []
y2 = []
y3 = []
y4 = []
Ts = 0.2
tMax = 10
f0 = 0.1
N = int(tMax/Ts) + 1

for i in range(N):

    t.append(i*Ts)
    y1.append(np.sin(2*np.pi*f0*t[-1]))
    y2.append(np.cos(2*np.pi*f0*t[-1]))
    y3.append(np.exp(alpha*t[-1]))
    y4.append(np.sqrt(t[-1]))

    # Set data

    l1.set_data(t,y1)
    l2.set_data(t,y2)
    l3.set_data(t,y3)
    l4.set_data(t,y4)
    clear_output(wait=True)  

    ax1.set_xlim([0, tMax])
    ax1.set_ylim([-1.2, 1.2])

    ax2.set_xlim([0, tMax])
    ax2.set_ylim(0, np.exp(alpha*tMax))

    ax3.set_xlim([0, tMax])
    ax3.set_ylim(0, np.sqrt(tMax))

    display(fig)  
    fig.tight_layout(pad = 2.0)
```
